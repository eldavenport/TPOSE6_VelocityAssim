```python
## Read in ITER 0 Diagnostics 

import xarray as xr
from xmitgcm import open_mdsdataset
import numpy as np
import warnings
# warnings.filterwarnings("ignore")

data_dir = '/data/SO3/edavenport/tpose6/sep2012/run_iter0/'
grid_dir = '/data/SO6/TPOSE_diags/tpose6/grid_6/'

num_diags = 122 #
itPerFile = 72 # 1 day
intervals = range(itPerFile,itPerFile*num_diags+1,itPerFile)

prefix = ['diag_state']
tpose_ds = open_mdsdataset(data_dir=data_dir,grid_dir=grid_dir,iters=intervals,prefix=prefix)
```

```python
import matplotlib.pyplot as plt
import cmocean.cm as cmo
fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[-1,0].plot(ax=ax[0], vmin=20, vmax=32, cmap=cmo.thermal)
tpose_ds.SALT[-1,0].plot(ax=ax[1], vmin=30, vmax=37, cmap=cmo.haline)
tpose_ds.UVEL[-1,0].plot(ax=ax[2], vmin=-1.5, vmax=1.5, cmap=cmo.balance)
tpose_ds.VVEL[-1,0].plot(ax=ax[3], vmin=-0.5, vmax=0.5, cmap=cmo.balance)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[:,0].mean(dim='time').plot(ax=ax[0], vmin=20, vmax=31, cmap=cmo.thermal)
tpose_ds.SALT[:,0].mean(dim='time').plot(ax=ax[1], vmin=30, vmax=37, cmap=cmo.haline)
tpose_ds.UVEL[:,0].mean(dim='time').plot(ax=ax[2], vmin=-1.5, vmax=1.5, cmap=cmo.balance)
tpose_ds.VVEL[:,0].mean(dim='time').plot(ax=ax[3], vmin=-0.5, vmax=0.5, cmap=cmo.balance)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[:,20].mean(dim='time').plot(ax=ax[0], vmin=12, vmax=29.25, cmap=cmo.thermal)
ax[0].set_title('Mean THETA at 75m')
tpose_ds.SALT[:,20].mean(dim='time').plot(ax=ax[1], vmin=33.0, vmax=36.5, cmap=cmo.haline)
ax[1].set_title('Mean SALT at 75m')
tpose_ds.UVEL[:,20].mean(dim='time').plot(ax=ax[2], vmin=-1.5, vmax=1.5, cmap=cmo.balance)
ax[2].set_title('Mean UVEL at 75m')
tpose_ds.VVEL[:,20].mean(dim='time').plot(ax=ax[3], vmin=-0.3, vmax=0.3, cmap=cmo.balance)
ax[3].set_title('Mean VVEL at 75m')
plt.tight_layout()
```

```python
lats = tpose_ds.YC.values
latui = np.argmin(np.abs(lats - 0.1))  # Upper latitude index
latli = np.argmin(np.abs(lats - -0.1))  # Lower latitude index

fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[-1,:40,latli:latui].mean(dim='YC').plot(ax=ax[0], vmin=10, vmax=31, cmap=cmo.thermal)
ax[0].set_title('Final THETA at lat 0.0')
tpose_ds.SALT[-1,:40,latli:latui].mean(dim='YC').plot(ax=ax[1], vmin=32.5, vmax=35.5, cmap=cmo.haline)
ax[1].set_title('Final SALT at lat 0.0')
tpose_ds.UVEL[-1,:40,latli:latui].mean(dim='YC').plot(ax=ax[2], vmin=-1.5, vmax=1.5, cmap=cmo.balance)
ax[2].set_title('Final UVEL at lat 0.0')
tpose_ds.VVEL[-1,:40,latli:latui].mean(dim='YG').plot(ax=ax[3], vmin=-0.6, vmax=0.6, cmap=cmo.balance)
ax[3].set_title('Final VVEL at lat 0.0')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[:,:40,latli:latui].mean(dim='YC').mean(dim='time').plot(ax=ax[0], vmin=10, vmax=31, cmap=cmo.thermal)
ax[0].set_title('Mean THETA at lat 0.0')
tpose_ds.SALT[:,:40,latli:latui].mean(dim='YC').mean(dim='time').plot(ax=ax[1], vmin=32.5, vmax=35.5, cmap=cmo.haline)
ax[1].set_title('Mean SALT at lat 0.0')
tpose_ds.UVEL[:,:40,latli:latui].mean(dim='YC').mean(dim='time').plot(ax=ax[2], vmin=-1.5, vmax=1.5, cmap=cmo.balance)
ax[2].set_title('Mean UVEL at lat 0.0')
tpose_ds.VVEL[:,:40,latli:latui].mean(dim='YG').mean(dim='time').plot(ax=ax[3], vmin=-0.6, vmax=0.6, cmap=cmo.balance)
ax[3].set_title('Mean VVEL at lat 0.0')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 15))
tpose_ds.THETA[:,:40,latli:latui].mean(dim='YC').var(dim='time').plot(ax=ax[0], vmin=0.0, vmax=2.0, cmap=cmo.thermal)
ax[0].set_title('Var THETA at lat 0.0')
tpose_ds.SALT[:,:40,latli:latui].mean(dim='YC').var(dim='time').plot(ax=ax[1], vmin=0.0, vmax=0.1, cmap=cmo.haline)
ax[1].set_title('Var SALT at lat 0.0')
tpose_ds.UVEL[:,:40,latli:latui].mean(dim='YC').var(dim='time').plot(ax=ax[2], vmin=0.0, vmax=0.15, cmap=cmo.balance)
ax[2].set_title('Var UVEL at lat 0.0')
tpose_ds.VVEL[:,:40,latli:latui].mean(dim='YG').var(dim='time').plot(ax=ax[3], vmin=0.0, vmax=0.2, cmap=cmo.balance)
ax[3].set_title('Var VVEL at lat 0.0')
plt.tight_layout()
```
